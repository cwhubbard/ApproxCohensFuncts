```python
import math
```

```python
def ApproxCohens(group1mean, group1SE, group1N, group2mean, group2SE, group2N, isSD):
 if isSD == 1: #if the values given do not need to be converted into standard deviation.
    if group1mean >= group2mean:
        g1SD = group1SE;
        g2SD = group2SE;
        differ_in_groups = (group1mean - group2mean);
        
        pooledSD = math.sqrt(((group1N-1)*g1SD**2 +(group2N-1)*g2SD**2)/(group1N+group2N-2));
        
        ApproxCohenD = differ_in_groups/pooledSD;
        print(ApproxCohenD)
    elif group1mean < group2mean:
        g1SD = group1SE;
        g2SD = group2SE;
        differ_in_groups = (group2mean - group1mean);
        pooledSD = math.sqrt(((group2N-1)*g2SD**2 +(group1N-1)*g1SD**2)/(group2N+group1N-2));
        
        ApproxCohenD = differ_in_groups/pooledSD;
        print(ApproxCohenD)
 elif isSD == 0:
    if group1mean >= group2mean:
        g1SD = group1SE * math.sqrt(group1N);#converts into SD
        g2SD = group2SE * math.sqrt(group2N);


        differ_in_groups = (group1mean - group2mean);
        pooledSD = math.sqrt(((group1N-1)*g1SD**2 +(group2N-1)*g2SD**2)/(group1N+group2N-2));

        ApproxCohenD = differ_in_groups/pooledSD;
        print(ApproxCohenD)
    elif group1mean < group2mean:
        g1SD = group1SE * math.sqrt(group1N);#converts into SD
        g2SD = group2SE * math.sqrt(group2N);


        differ_in_groups = (group2mean - group1mean);
        pooledSD = math.sqrt(((group2N-1)*g2SD**2 +(group1N-1)*g1SD**2)/(group2N+group1N-2));
        ApproxCohenD = differ_in_groups/pooledSD;
        print(ApproxCohenD)

```

```python
ApproxCohens(3,0.5,20,0,0.5,20,1) #checked against matlab version yields same value
ApproxCohens(3,0.5,20,4,0.5,20,1) 
ApproxCohens(3,0.5,20,0,0.5,20,0)
ApproxCohens(3,0.5,20,4,0.5,20,0) 
```
